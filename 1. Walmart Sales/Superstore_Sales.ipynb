```python
missing_values = superstore_df_csv.isnull().sum()
missing_percentage = (missing_values / len(superstore_df_csv)) * 100

# Combine into a DataFrame
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage (%)': missing_percentage})

# Print the result
print(missing_data)
```
