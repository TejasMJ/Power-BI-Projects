```python
import pandas as pd
```

```python
hotel_df = pd.read_csv('hotel_bookings.csv')
```

```python
missing_values = hotel_df.isnull().sum()
missing_percentage = (missing_values / len(hotel_df)) * 100

missing_data = pd.DataFrame({'Missing_Values': missing_values, 'Percentage (%)': missing_percentage})
missing_data
```

```python
hotel_df = hotel_df.drop(columns=['company','agent','adr','assigned_room_type', 'previous_cancellations', 'previous_bookings_not_canceled', 'lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights'])
```

```python
hotel_df['country'] = hotel_df['country'].fillna('Other')
hotel_df['children'] = hotel_df['children'].fillna(0)
```

```python
hotel_df.dtypes
```

```python
hotel_df['Total Guests'] = hotel_df['adults'] + hotel_df['children'] + hotel_df['babies']
```

```python
hotel_df.columns
```

```python
hotel_df.shape
```

```python
hotel_df["date"] = pd.to_datetime(hotel_df["arrival_date_year"].astype(str) + '-' + hotel_df["arrival_date_month"].astype(str) + '-' + hotel_df["arrival_date_day_of_month"].astype(str))
```

```python
hotel_df = hotel_df.drop(columns=['reservation_status_date', 'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'])
```
