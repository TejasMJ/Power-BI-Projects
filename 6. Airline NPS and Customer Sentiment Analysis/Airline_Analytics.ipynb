```python
airline_df[["Seat Comfort", "Cabin Staff Service", "Food & Beverages", "Ground Service", 'Inflight Entertainment', 'Value For Money']] = airline_df[["Seat Comfort", "Cabin Staff Service", "Food & Beverages", "Ground Service", 'Inflight Entertainment', 'Value For Money']].fillna(1)
```

```python
airline_df[["Seat Comfort", "Cabin Staff Service", "Food & Beverages", "Ground Service", 'Inflight Entertainment', 'Value For Money']] = airline_df[["Seat Comfort", "Cabin Staff Service", "Food & Beverages", "Ground Service", 'Inflight Entertainment', 'Value For Money']].astype(int)
```

```python
missing_values = airline_df.isnull().sum()
missing_percentage = (airline_df.isnull().sum() / len(airline_df)) * 100

missing_data = pd.DataFrame({'Missing_Values': missing_values, 'Percentage (%)': missing_percentage})
missing_data
```

```python
airline_df = airline_df.drop(columns=['Route'])
```

```python
airline_df["Type Of Traveller"] = airline_df["Type Of Traveller"].fillna('Solo Leisure')
```

```python
airline_df["Seat Type"] = airline_df["Seat Type"].fillna('Economy Class')
```

```python
missing_values = airline_df.isnull().sum()
missing_percentage = (airline_df.isnull().sum() / len(airline_df)) * 100

missing_data = pd.DataFrame({'Missing_Values': missing_values, 'Percentage (%)': missing_percentage})
missing_data
```

```python
airline_df.dropna(subset=["Date Flown"], inplace=True)
```
